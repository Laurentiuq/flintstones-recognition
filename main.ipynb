```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
```

```python
personaje = ["barney", "betty", "fred", "wilma"]

nr = 0
for pers in personaje:
    pers_dir = os.path.join("antrenare", pers)
    pers_annotation = os.path.join("antrenare", pers + "_annotations.txt")

    annotations = open(pers_annotation, "r")
    img_annotation = annotations.readline()

    while annotations:
        if img_annotation == "":
            break
        img_name = img_annotation.split()[0]
        img_name_last = img_name
        img_path = os.path.join(pers_dir, img_name)
        img = cv.imread(img_path)


        while img_name == img_name_last:

            x_min = int(img_annotation.split()[1])
            x_max = int(img_annotation.split()[3])
            y_min = int(img_annotation.split()[2])
            y_max = int(img_annotation.split()[4])

            img_fata_croped = img[y_min:y_max, x_min:x_max]
            img_fata_croped = cv.resize(img_fata_croped, (64, 64))
            cv.imwrite(os.path.join("pozitive_total", str(nr).zfill(5) + ".jpg"), img_fata_croped)



            img[y_min:y_max, x_min:x_max] = cv.GaussianBlur(img[y_min:y_max, x_min:x_max], (43,43), 50)


            img_name_last = img_name
            img_annotation = annotations.readline()
            if img_annotation == "":
                break
            img_name = img_annotation.split()[0]
            nr += 1
        
        cv.imwrite(os.path.join("negative_total", str(nr).zfill(5) + ".jpg"), img)
        print(f"Procesat imaginea {img_name}")

        # plt.imshow(img)
        # plt.title(img_path)
        # plt.show()
```

```python
"""Creeaza fisierul cu anotarile corespunzatoare"""
personaje = ["barney", "betty", "fred", "wilma"]

nr = 0

annotation_file = open("annotations_total.txt", "a")

for pers in personaje:
    pers_dir = os.path.join("antrenare", pers)
    pers_annotation = os.path.join("antrenare", pers + "_annotations.txt")

    annotations = open(pers_annotation, "r")
    img_annotation = annotations.readline()

    while annotations:
        if img_annotation == "":
            break

        img_name = img_annotation.split()[0]
        img_name_last = img_name
        img_path = os.path.join(pers_dir, img_name)
        img = cv.imread(img_path)


        while img_name == img_name_last:

            x_min = int(img_annotation.split()[1])
            x_max = int(img_annotation.split()[3])
            y_min = int(img_annotation.split()[2])
            y_max = int(img_annotation.split()[4])

            img_name_last = img_name
            img_annotation = annotations.readline()
            if img_annotation == "":
                break
            img_name = img_annotation.split()[0]
            nr += 1
            annotation_file.write(f"{str(nr).zfill(5)}.jpg {x_min} {y_min} {x_max} {y_max}\n")

        # else:            
        # annotation_file.write(f"{str(nr).zfill(5)}.jpg {x_min} {y_min} {x_max} {y_max}\n")
        
        
        print(f"Procesat imaginea {img_name}")

        # plt.imshow(img)
        # plt.title(img_path)
        # plt.show()
annotation_file.close()
```
